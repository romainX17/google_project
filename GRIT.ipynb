```python
class Speaker():
    
    def __init__(self,arr):
        self.take_input = True
        self.is_finished = False
        self.model = NERTransformer.load_from_checkpoint("output/checkpointepoch=4.ckpt").cuda()
        self.arr = arr
        self.roles = ["painter_name","birth","death","nationality"]
        self.d = {"painter_name":1,"birth":2,"death":3,"nationality":4}
        
        
    def analyse_input(self,inputs):
        with open("data/test.json",'w',encoding='utf-8') as f:
            d = {}
            d["docid"] = "pred-1"
            d['doctext'] = inputs
            s = {}
            s['birth'] = s["death"] = s["nationality"] = s["painter_name"] = []
            #s["painter_name"] = [[[text[:i-1],0]]]
            d["extracts"] = s
            json.dump(d,f,ensure_ascii=False)
        model = self.model
        examples = read_examples_from_file("data","test",model.tokenizer)
        features = convert_examples_to_features(
                    examples,
                    # self.labels,
                    model.hparams.max_seq_length_src,
                    model.hparams.max_seq_length_tgt,
                    model.tokenizer,
                    cls_token=model.tokenizer.cls_token,
                    sep_token=model.tokenizer.sep_token,
                    pad_token=model.tokenizer.convert_tokens_to_ids([model.tokenizer.pad_token])[0],
                    pad_token_label_id=model.pad_token_label_id,
                )        
        batch = [] 
        batch.append(torch.tensor([features[0].input_ids]).cuda())#to(model.device))
        batch.append(torch.tensor([features[0].input_mask]).to(model.device))
        batch.append(torch.tensor([features[0].segment_ids]).to(model.device))
        batch.append(torch.tensor([features[0].position_ids]).to(model.device))
        batch.append(torch.tensor([features[0].label_ids]).to(model.device))
        batch.append(torch.tensor([features[0].docid]).to(model.device))
        x = model.test_step(batch,0)
        role_list = ["painter_name", "birth", "death", "nationality"]
        p_extract = x['pred_extract'][0]
        l = {}
        for idx, role in enumerate(role_list):
            l[role] = []
            if idx+1 > len(p_extract): 
                continue
            elif p_extract[idx]:
                l[role] = p_extract[idx]
        return l
        
    def find_id(self,l):
        roles = self.roles
        arr = self.arr
        if l["painter_name"]:
            i = np.where([l["painter_name"][0][0] == i.lower() for i in arr[:,1]])[0]
        else:
            i = []
        if len(i)>0:
            return [i[0]],True
        else:
            inter = []
            for r in roles:
                u = list(self.find_intersect(l,r))
                if inter and u:
                    inter = list(np.intersect1d(inter,u))
                elif u:
                    inter = u
            return list(arr[inter,0]), False
    
    def find_intersect(self,l,role):
        d = self.d
        arr = self.arr
        if not l[role]:
            return []
        if role == "painter_name":
            infos = arr[np.where([l[role][0][0] in i.lower() for i in arr[:,d[role]]])[0],0]
        elif role == "nationality":
            infos = arr[np.where([l[role][0][0] in [j.strip("'") for j in i.strip('][').split(", ")] for i in arr[:,d[role]]])[0],0]
        else:
            infos = arr[np.where(int(l[role][0][0]) == arr[:,d[role]]+np.isnan(arr[:,d[role]].astype(float)))[0],0]
        return infos.astype(int)
    
    def compare_with_base(self,enforce):
        roles = self.roles
        arr = self.arr
        d = self.d
        l = self.l
        if enforce:
            ids, confidence = [self.id], True
        else:
            ids, confidence = self.find_id(l)
            
        if confidence:
            vp = arr[ids,2:]
            
            mismaches = [False,False,False]
            new_ones = [False,False,False]
            u = [None,None,None]
            v = [None,None,None]
            for i in range(1,len(roles)):
                role = roles[i]
                if role == "nationality":
                    pred = l[role][0][0] if l[role] else None
                    labels = vp[0][2].strip("'][").split(", ")
                    u[-1] = pred
                    v[-1] = labels
                    if not pred is None and not pred in [i.lower() for i in labels]:
                        mismaches[-1] = True
                else:
                    pred = int(l[role][0][0]) if l[role] else None
                    labels = int(vp[0][d[role]-2])
                    u[d[role]-2] = pred
                    v[d[role]-2] = labels
                    if not pred is None and not pred == labels:
                        mismaches[d[role]-2] = True
                    
            #u = [l[role][0][0] if l[role] else None for role in roles[1:]]
            #u = [int(u[i]) if (i in [0,1] and not u[i] is None) else u[i] for i in range(3)]
            #v = arr[ids,2:]
            #v = [int(v[0][0]),int(v[0][1]),v[0][2].strip("'][,").lower()]
            #mismaches =  (np.array(u) == v)+ np.array([i is None for i in u])
            #mismaches = [not i for i in mismaches]
            nl = []
            for j in range(len(mismaches)):
                if mismaches[j]:
                    if j == 2:
                        s = "For the "+roles[j+1]+", the value we identify is different from the one in our data base.\nYours is "+str(u[j])+" while ours is "+" or ".join(v[j])+".\n"
                    else:    
                        s = "For the "+roles[j+1]+", the value we identify is different from the one in our data base.\nYours is "+str(u[j])+" while ours is "+str(v[j])+".\n"
                    nl.append(s)
            return confidence, nl
        
        else:
            return confidence, ids
    
    def main(self):
        while not self.is_finished:
            if self.take_input:
                ins = input("Text\n")
                if ins == "stop":
                    break
                self.l = self.analyse_input(ins)
            c, u = self.compare_with_base(not self.take_input)
            if c:
                if u:
                    for s in u:
                        input(s+"Replace ?(y/n)\n")
                else:
                    print("Your data seems aligned to ours")
                self.take_input = True
            else:
                if len(u) == 1:
                    ins = input("We think you meant "+arr[u[0],1]+", is that correct ?(y/n)")
                    if ins == 'y':
                        self.id = u[0]
                        self.take_input = False
                elif len(u)>1 and len(u)<5:
                    ins = input("We have found "+str(len(u))+" matching possiblities would you like to check them?(y/n)\n")
                    if ins == 'y':
                        for ids in u:
                            ins = input("Is it "+arr[ids,1]+"?(y/n)")
                            if ins == 'y':
                                self.id = ids
                                self.take_input = False
                                break
                elif len(u)>=5:
                    print("Sadly we have to much corresponding possible matches!")
                else:
                    print("Sadly no match!")
                    if self.l["painter_name"]:
                        s = []
                        for role in self.roles[1:]:
                            if self.l[role]:
                                s.append(self.l[role][0][0]+" for the attribute "+role)
                        if len(s)==0:
                            s =' '
                        elif len(s)==1:
                            s= " with "+s[0]
                        else:
                            u = " with "
                            for i in range(len(s)):
                                if i == len(s) - 2:
                                    u = u + s[i] + " and "
                                else:
                                    u = u + s[i] + ", "
                            s = u
                        input("Do you want to add "+self.l["painter_name"][0][0]+s+"to the data base?(y/n)\n")
    
```
